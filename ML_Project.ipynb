Data Preprocessing

```python
yield_df.head()
```

Encoding Categorical Variables:

```python
from sklearn.preprocessing import OneHotEncoder
final_df_onehot = pd.get_dummies(yield_df, columns=['Area',"Item"], prefix = ['Country',"Item"])
features=final_df_onehot.loc[:, final_df_onehot.columns != 'hg/ha_yield']
label=yield_df['hg/ha_yield']
features.head()
```

```python
features = features.drop(['Year'], axis=1)
```

```python
features.info()
features.head()
```

Scaling Features:

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
features=scaler.fit_transform(features)
```

```python
features
```

Training Data:

```python
from sklearn.model_selection import train_test_split
train_data, test_data, train_labels, test_labels = train_test_split(features, label, test_size=0.2, random_state=42)
```

Model Comparison & Selection

```python
from sklearn.metrics import r2_score
r2=[]
def compare_models(model):
     model_name = model.__class__.__name__
     fit=model.fit(train_data,train_labels)
     y_pred=fit.predict(test_data)
     r2=r2_score(test_labels,y_pred)
     return([model_name,r2])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn import svm
model_names = ['Gradient Boosting', 'Random Forest', 'Decision Tree', 'Support Vector Machine']
models = [
    GradientBoostingRegressor(n_estimators=200, max_depth=3, random_state=0),
    RandomForestRegressor(n_estimators=200, max_depth=3, random_state=0),
    DecisionTreeRegressor(),
    svm.SVR()]
```

```python
model_train=list(map(compare_models,models))
```

```python
print(*model_train, sep = "\n")
```

```python
from sklearn.metrics import mean_squared_error
r2_scores = []

# Train and evaluate each model
for model in models:
     fit2=model.fit(train_data,train_labels)
     y_pred=fit2.predict(test_data)
     r2=r2_score(test_labels,y_pred)
     r2_scores.append(r2)

# Create a bar plot to compare MSE scores of different models
plt.figure(figsize=(10, 6))
plt.bar(model_names, r2_scores, color='skyblue')
plt.title('R2 of Regression Models')
plt.ylabel('R2_Score')
plt.xticks(rotation=45)
plt.show()
```

```python
final_df_onehot = final_df_onehot.drop(['Year'], axis=1)
final_df_onehot.head()
```

```python
#setting test data to columns from dataframe and excluding 'hg/ha_yield' values where ML model should be predicting

df_test=pd.DataFrame(test_data,columns=final_df_onehot.loc[:, final_df_onehot.columns != 'hg/ha_yield'].columns)

# using stack function to return a reshaped DataFrame by pivoting the columns of the current dataframe

country_name=df_test[[col for col in df_test.columns if 'Country' in col]].stack()[df_test[[col for col in df_test.columns if 'Country' in col]].stack()>0]
country_list=list(pd.DataFrame(country_name).index.get_level_values(1))
countries=[i.split("_")[1] for i in country_list]
item_name=df_test[[col for col in df_test.columns if 'Item' in col]].stack()[df_test[[col for col in df_test.columns if 'Item' in col]].stack()>0]
item_list=list(pd.DataFrame(item_name).index.get_level_values(1))
items=[i.split("_")[1] for i in item_list]
```

```python
df_test.head()
```

```python
df_test.drop([col for col in df_test.columns if 'Item' in col],axis=1,inplace=True)
df_test.drop([col for col in df_test.columns if 'Country' in col],axis=1,inplace=True)
df_test.head()
```

```python
df_test
```

```python
df_test['Country']=countries
df_test['Item']=items
df_test.head()
```

```python
clf=DecisionTreeRegressor()
test_model=clf.fit(train_data,train_labels)

df_test["prediction_yield"]= test_model.predict(test_data)
df_test["actual_yield"]=pd.DataFrame(test_labels)["hg/ha_yield"].tolist()
test_group=df_test.groupby("Item")
test_group.apply(lambda x: r2_score(x.actual_yield,x.prediction_yield))
```

```python
fig, axis = plt.subplots()

axis.scatter(df_test["actual_yield"], df_test["prediction_yield"],edgecolors=(0, 0, 0))

axis.set_xlabel('Actual')
axis.set_ylabel('Predicted')
axis.set_title("Actual vs Predicted")
plt.show()
```

```python
def adjusted_r_squared(y,yhat,x):
    score=1 - (((1-(r2_score(y,yhat)))*(len(y)-1))/(len(y)-x.shape[1]-2))
    return score

test_group.apply(lambda x: adjusted_r_squared(x.actual_yield,x.prediction_yield,x))
```
